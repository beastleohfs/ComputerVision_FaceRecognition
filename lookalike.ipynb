```python
import face_recognition
from pathlib import Path
from PIL import Image
```

```python
image=face_recognition.load_image_file("test_face.jpg")
```

```python
image_encoding=face_recognition.face_encodings(image)[0]
```

```python
best_face_distance=1.0
best_face_image=None
```

```python
for image_path in Path("people").glob("*.png"):
    unknown=face_recognition.load_image_file(image_path)
    face_encodings=face_recognition.face_encodings(unknown)
    face_distance=face_recognition.face_distance(face_encodings,image_encoding)[0]
    if face_distance < best_face_distance:
        best_face_distance=face_distance
        best_face_image=unknown
```

```python
import matplotlib.pyplot as plt
```

```python
pil_image=Image.fromarray(best_face_image)
plt.imshow(pil_image)
```
